## 1. Setup

````python
!pip install pandas matplotlib seaborn numpy -q
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use("seaborn-v0_8-whitegrid")
print("✅ Ready")
````

## 2. Load sample data

````python
# Download from Zenodo: https://github.com/juanmoisesd/venezuela-economic-impact-database-indicators-2024-2025
np.random.seed(42)
countries = ["Brazil","Mexico","Argentina","Colombia","Chile","Peru","Venezuela","Ecuador"]
years = list(range(2015, 2026))
rows = []
for c in countries:
    base = np.random.uniform(8,25)
    for i,y in enumerate(years):
        rows.append({"country":c,"year":y,"value":round(base+i*np.random.uniform(0.1,0.8)+np.random.normal(0,0.5),2),"n":int(np.random.uniform(500,5000))})
df = pd.DataFrame(rows)
print(f"Shape: {df.shape}")
df.head()
````

## 3. Visualization

````python
fig,axes=plt.subplots(1,2,figsize=(14,5))
for c in countries:
    s=df[df["country"]==c]
    axes[0].plot(s["year"],s["value"],marker="o",ms=3,label=c)
axes[0].set_title("Trend by Country",fontsize=13,fontweight="bold")
axes[0].legend(fontsize=7,ncol=2)
pivot=df.pivot_table(values="value",index="country",columns="year",aggfunc="mean").round(1)
sns.heatmap(pivot,ax=axes[1],cmap="YlOrRd",annot=True,fmt=".1f",annot_kws={"size":7},linewidths=0.5)
axes[1].set_title("Heatmap",fontsize=13,fontweight="bold")
plt.tight_layout()
plt.savefig("example_plot.png",dpi=150,bbox_inches="tight")
plt.show()
````
